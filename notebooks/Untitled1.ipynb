```python
import pandas as pd
from sqlalchemy import create_engine

# Load cleaned data from the correct location
df = pd.read_csv(
    r"D:/AKRAM/data analysis project/ecommerce_customer_behavior_project/data/processed/customers_clean.csv"
)

# Create connection to MySQL
engine = create_engine(
    "mysql+mysqlconnector://root:56600@localhost:3306/ecommerce_db"
)

# Upload DataFrame to MySQL table
df.to_sql("customers", con=engine, if_exists="replace", index=False)

print("Data uploaded to MySQL successfully!")
print(f"Rows uploaded: {len(df)}")
```
